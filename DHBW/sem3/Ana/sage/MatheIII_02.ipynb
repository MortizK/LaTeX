# 1.3 Stetigkeit

1.) Beispielfunktion 
$$ f(x_1,x_2) = \begin{cases}
    \frac{2 x_1 x_2}{x_1^2 + x_2^2}     & \text{ falls } (x_1,x_2) \neq (0,0) \\
    \;\;0  & \text{ falls } x_1 = x_2 = 0
\end{cases}$$

```python
x, y = var('x,y')
F = plot3d(2 * x * y / (x^2 + y^2), (x,-2,2), (y,-2,2), plot_points = 400)
t = var('t')
HL = parametric_plot3d((t, t, 1), (-2,2), color="red", thickness = 8)
HX = parametric_plot3d((t, 0, 0), (-2,2), color="orange", thickness = 8)
HY = parametric_plot3d((0, t, 0), (-2,2), color="yellow", thickness = 8)
show(F + HL + HX + HY)
```

2.) Beispielfunktion 
$$
f(x_1,x_2) = \begin{cases}
    x_1 * \sin(\frac{1}{x_1^2 + x_2^2})     & \text{ falls } (x_1,x_2) \neq (0,0) \\
    \;\;0  & \text{ falls } x_1 = x_2 = 0
\end{cases}
$$

```python
x, y = var('x,y')
plot3d(x * sin(1/(x^2 + y^2)), (x,-1,1), (y,-1,1),plot_points = 400)
```
